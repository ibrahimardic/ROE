# **Baseline Model**


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.svm import SVC
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# @title Features and Target Values
X_baseline = df.drop(columns = {'Room_Occupancy_Count', 'Time', 'Date' }, axis=1)  # Select all columns except "Room_Occupancy_Count" and time datas
y_baseline = df['Room_Occupancy_Count']  # Select the target column#
```

```python
# @title Train Test Split
X_train_bl, X_test_bl, y_train_bl, y_test_bl = train_test_split(X_baseline, y_baseline, test_size=0.2, random_state=13, stratify = y_baseline)
X_train_bl, X_val_bl, y_train_bl, y_val_bl = train_test_split(X_train_bl, y_train_bl, test_size=0.2, random_state=13, stratify = y_train_bl)
```

```python
# @title Random Classifier

class RandomClassifier:

    def __init__(self, num_classes):
        self.num_classes = num_classes

    def predict(self, X_baseline):
        return np.random.randint(low=0, high=self.num_classes, size=X_baseline.shape[0])

def baseline_accuracy(y_true, y_pred):
    correct = np.sum(y_true == y_pred)
    total = len(y_true)
    return correct / total

def baseline_evaluate(y_true, y_pred):
    acc = round((baseline_accuracy(y_true, y_pred))*100,2)
    precision = round((precision_score(y_true, y_pred, average='weighted'))*100,2)
    recall = round((recall_score(y_true, y_pred, average='weighted'))*100,2)
    f1 = round((f1_score(y_true, y_pred, average='weighted'))*100,2)

    print(f"Accuracy : {acc}")
    print(f"Precision : {precision}")
    print(f"Recall : {recall}")
    print(f"F1 Score : {f1}")

random_model = RandomClassifier(4)

y_true = y_test_bl.values
y_pred_random = random_model.predict(X_test_bl.values)

baseline_evaluate(y_true, y_pred_random)
```

```python
# @title Dummy Classifier
def dummy_evaluate(y_true, y_pred):
    acc = round((baseline_accuracy(y_true, y_pred))*100,2)
    precision = round((precision_score(y_true, y_pred, average='weighted'))*100,2)
    recall = round((recall_score(y_true, y_pred, average='weighted'))*100,2)
    f1 = round((f1_score(y_true, y_pred, average='weighted'))*100,2)

    print(f"Accuracy : {acc}")
    print(f"Precision : {precision}")
    print(f"Recall : {recall}")
    print(f"F1 Score : {f1}")

dummy_model = DummyClassifier(strategy='stratified', random_state=42)
dummy_model.fit(X_train_bl.values, y_train_bl.values)

y_pred_dummy = dummy_model.predict(X_test_bl.values)

dummy_evaluate(y_true, y_pred_dummy)
```

```python
scores_dummy = cross_val_score(dummy_model, X_train_bl, y_train_bl, cv=5)

print("Cross-validation scores (Dummy Classifier):", scores_dummy)
print("Mean cross-validation score (Dummy Classifier):", np.mean(scores_dummy))
```

```python
clf_uniform = DummyClassifier(strategy='uniform')
clf_uniform.fit(X_train_bl, y_train_bl)

y_pred_uniform = clf_uniform.predict(X_test_bl)

accuracy_uniform = accuracy_score(y_test_bl, y_pred_uniform)
print("Accuracy (Uniform Rule):", accuracy_uniform)
```

```python
scores_clf = cross_val_score(clf_uniform, X_train_bl, y_train_bl, cv=5)

print("Cross-validation scores (Random Classifier):", scores_clf)
print("Mean cross-validation score (Random Classifier):", np.mean(scores_clf))
```

# **Data Manipulation**

```python
df1= df.copy()
df1['Date'] = pd.to_datetime(df1['Date'])
df1['Time'] = pd.to_datetime(df1['Time'], format='%H:%M:%S')
```

```python
# @title Merging S1_Temp and S3_Temp using their means
df1['Avg_Temp'] = (df1['S1_Temp'] + df1['S3_Temp'] ) / 2
column_to_drop = ['S3_Temp']

df1.drop(column_to_drop, axis=1, inplace = True)

df1['S1_Temp'] = df1['Avg_Temp']
df1.drop(columns = 'Avg_Temp', inplace = True)
```

```python
# @title Merging PIR sensors

df1['PIR'] = (df1['S6_PIR'] + df1['S7_PIR']) >= 1
df1['PIR'] = (df1['PIR'] >= 1).astype(int)

df1.drop(['S7_PIR'], axis=1, inplace = True)
```

```python
df1['S6_PIR'] = df1['PIR']
```

```python
df1.drop('PIR', axis=1, inplace = True)
```

```python
df1.rename(columns = {'S1_Temp': 'Avg_Temp(S1-S3)'}, inplace = True)
df1
```

```python
# @title Implementing Date Parts

df1['Day_Parts'] = df1['Time'].dt.hour.apply(lambda x: 1 if x >= 6 and x < 11 else 2 if x >= 11 and x < 13 else 3 if x >=13 and x < 18 else 4 if x >= 18 and x < 23 else 5)
df1['Day_Parts'].value_counts()
```

```python
plt.figure(figsize=(12,7))
ax = sns.countplot(data = df1,x = 'Room_Occupancy_Count', hue = 'Day_Parts', hue_order = [1,2,3,4,5])
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```

1 -> Morning (6AM to 11AM)

2 -> Noon (11AM to 1PM)

3 -> Afternoon (1PM to 6PM)

4 -> Evening (6PM to 11PM)

5 -> Night (11PM to 6AM)
